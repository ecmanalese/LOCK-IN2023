```python
import pandas as pd
import requests
import bs4

statURL = 'https://liquipedia.net/valorant/VCT/2023/LOCK_IN_S%C3%A3o_Paulo/Statistics'

response = requests.get(statURL)
soup = bs4.BeautifulSoup(response.text, 'html.parser')
table = soup.find('table')

df=pd.read_html(str(table))
df=pd.DataFrame(df[0])
df
```

```python
#Get each data through table rows
table_data = table.find_all('tr')

#separate columns through table data
headers = []
for td in table.find_all('th'):
    headers.append(td.text)

headers
```

```python
#data and compstat will be merged to create finalized table
data={'Player':[],
      'Country':[],
      'Team':[],
      'Agents':[]}

CompStat = []

#gets index numbers of rows
for i in range(1, len(table_data)):

    #data on Country, Team, and Agents
    null = []

    #data already available on table
    IncStat = []

    #gets stat values of each player
    #no values for country, team, and agents as they are shown as images
    for td in table_data[i].find_all('td'):

        #country is at the start of each player's name, replaced by whitespace, thus need for removal
        IncStat.append(td.get_text(strip=True))

        #gets data on Country, Team, and Agents
        for x in td.find_all('a'):
            null.append(x.get('title'))

    #removes Team duplicate
    del null[3]

    #removes null values
    del IncStat[1:3]

    #appends values for the columns
    data['Country'].append(null[0])
    data['Player'].append(null[1])
    data['Team'].append(null[2])
    data['Agents'].append(null[3:])
    CompStat.append(IncStat)

print(data)
print(CompStat)
```

```python
stat = pd.DataFrame(CompStat)
stat.columns=['Player', 'Maps', 'K', 'D', 'A', 'KD', 'KDA', 'ACS/Map', 'K/Map', 'D/Map', 'A/Map']
playerinfo = pd.DataFrame(data)

finalstats = pd.merge(playerinfo, stat, how='right')
finalstats
```

```python
#Maps Picked

map_table = soup.find('table',attrs={'class': 'table table-bordered'})
map_values = map_table.find_all('tr')

map_values
```

```python
#Side Stats
side_table = soup.find('table',attrs={'class': 'wikitable wikitable-bordered wikitable-striped'})
side_values = side_table.find_all('tr')

side_values

final_data=[]
for i in range(len(side_values)):
    temp=[]
    for td in side_values[i].find_all(['td','th']):
        temp.append(td.text)
    final_data.append(temp)
side_pick = pd.DataFrame(final_data)

side_pick.columns = side_pick.iloc[0]
side_pick.drop(0,inplace=True)
side_pick.set_index(keys='Map',inplace=True)

#separating numbers and percentage values to their own columns
side_pick[['Atk Wins', 'Atk Win %']] = side_pick['Atk Wins'].apply(lambda x: pd.Series(str(x).split(' ')))
side_pick[['Def Wins', 'Def Win %']] = side_pick['Def Wins'].apply(lambda x: pd.Series(str(x).split(' ')))

#changing values  for quantitative processing
side_pick['Atk Win %'] = side_pick['Atk Win %'].str.replace(r"\(|\)|%", '', regex=True)
side_pick['Def Win %'] = side_pick['Def Win %'].str.replace(r"\(|\)|%", '', regex=True)
side_pick = side_pick[['Atk Wins', 'Atk Win %', 'Def Wins', 'Def Win %']].apply(pd.to_numeric)

side_pick.info()
```
